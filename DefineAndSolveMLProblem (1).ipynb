## Cleaning the data

```python
#Check for all the columns with missing data
nan_count = np.sum(df.isnull(), axis = 0)
nan_count
```

```python
#Create a list of all the columns with missing values

condition = nan_count != 0
col_names = nan_count[condition].index

nan_cols = list(col_names)
print("Columns with missing values: ", nan_cols)
```

```python
#Create a column for each in nan_cols with booleans that say whether or not the value is null
nan_exists_column = []
for col in nan_cols:
    df[col+'_na'] = df[col].isnull()
    nan_exists_column.append(col+'_na')
    

df.head()

#For test to see the NaN value change
print(df['GINI index (World Bank estimate), average 2000-15'][0])
```

```python
#Replace missing values in all numerical column with a mean of the other values in the column
for col in nan_cols:
    mean_vals = df[col].mean()
    df[col].fillna(value=mean_vals, inplace=True)

#Check to make sure all NaN values are replaced
for col in nan_cols:
    print(col, np.sum(df[col].isnull(), axis = 0))

print(df['GINI index (World Bank estimate), average 2000-15'][0])
```

```python
#Remove all the boolean columns
print(df.columns)

for nan in nan_exists_column:
    df.drop(columns = nan, inplace=True)

print(df.columns)
df.head()
```

## Using K-Fold Cross Validation 

```python
model2 = LinearRegression()

num_folds = 5
folds = KFold(n_splits = num_folds, random_state=None)

mse_scores = []
r2_scores = []

for train_row_index, test_row_index in folds.split(X_train):
    X_train_new  = X_train.iloc[train_row_index] 
    X_val = X_train.iloc[test_row_index]

    y_train_new = y_train.iloc[train_row_index]
    y_val = y_train.iloc[test_row_index]

    model2.fit(X_train_new, y_train_new)
    predictions = model.predict(X_val)

    mse = mean_squared_error(y_val, predictions)
    r2 = r2_score(y_val, predictions)

    mse_scores.append(mse)
    r2_scores.append(r2)

for i in range(len(mse_scores)):
    print('MSE score for iteration {0}: {1}'.format(i+1, mse_scores[i]))

for i in range(len(r2_scores)):
    print('R Squared score for iteration {0}: {1}'.format(i+1, r2_scores[i]))

mse_avg = sum(mse_scores)/num_folds
print('\nAverage MSE score: {}'.format(mse_avg))

r2_avg = sum(r2_scores)/num_folds
print('\nAverage MSE score: {}'.format(r2_avg))
    
```

```python
cv_scores = cross_val_score(model2, X_train, y_train, cv=5)
print(cv_scores)
```

Using K-Fold validation techniques also led to pretty similar results as it was without it. The model still seems to be performing pretty well even when it is being tested and validated on not seen before data, which seems to be a good indicator of its performance. Overall, it has a low RSME value and a high R^2 value on average, which depicts that it is likely a strong model.
